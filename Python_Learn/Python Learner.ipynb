## The upward datatype is a dictionary where first element is the key and the second element is the value stored in the key

```python
L = [ 1, 2, 4 , -5, 7 , 9 ,3,2]

for j in range(len(L)):
    m = L[j]
    idx = j
    c = j
    for i in range(j,len(L)):
        """
        9 3 4 1

        C = 0 
        9 < 9 NO
        C = 1
        
        3 4 1

        3 < 9
        m = 3 
        idx = 1
        c = 1

        4 1 
        4 < 3  NO
        c = 2 

        1
        1 < 3 YES
        m = 1
        idx = 2
        C = 3
        
        C increases every iteration 
        
        """
        if L[i] < m:
            m = L[i]
            idx = c
        c+= 1
    tmp = L[j]
    L[j] = m
    L[idx] = tmp 
print(L) 

```

```python
L = [ 1, 2, 4 , -5, 7 , 9 ,3,2]

for j in range(len(L)):
    c = j
    m = L[j]
    idx = j

    for i in range(j,len(L)):
        if L[i] < m:
            m = L[i]
            idx = c
        c += 1
    tmp = L[j]
    L[j] = m
    L[idx] = tmp

print (L)
```

```python
def findMin(L):
    min = L[0]
    idx = 0
    i = 0 
    for x in L:
        if x < min:
            min = x 
            idx = i
        i += 1 
    return min, idx
```

```python
a , b = findMin([1,2,3,4])
```

```python
print (a,b)
```

```python
def swapValues( L, idx1, idx2):
    temp = L[idx1] 
    L[idx1] = L[idx2]
    L[idx2] = temp
    return L
    
```

```python
swapValues( [1,2,3,4] , 0 , 1)
```
