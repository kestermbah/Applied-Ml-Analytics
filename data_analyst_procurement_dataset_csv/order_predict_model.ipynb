

##### The Big Three — Numeric Features Dominate

| Feature | Importance | Insight |
|---|---|---|
| `LineAmount` | 23% | High value orders are the strongest signal |
| `Quantity` | 22% | How many units ordered matters almost as much |
| `UnitPrice` | 21% | Price per unit is nearly equally important |

---

##### Time Features — Matter but Less

| Feature | Importance | Insight |
|---|---|---|
| `Month` | 12% | Seasonality plays a role — certain months are worse |
| `DayOfWeek` | 10% | Day of week has some influence |
| `Year` | 4% | Minimal impact |

---

##### 🏷️ Category Features — Barely Matter (~1–2% each)

All `ItemCategory_*` columns rank at the bottom with very low importance.

> Whether an order is **Software, Hardware, Services, Equipment, or Consumables** does not strongly predict lateness.

---

###### 💡 Key Takeaway

> **Large, expensive, high-quantity orders are most likely to run late.**
>
> Any order where both `LineAmount` and `Quantity` are high should be flagged as a lateness risk **before it ships**.
